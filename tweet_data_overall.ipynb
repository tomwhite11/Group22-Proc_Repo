```python
import os
import json
from pprint import pprint
import pandas as pd
```

```python
JSON_FILES= []

JSON_FILES.append("C:/Users/user/OneDrive - Cardiff University/en/2022-07-14.json/2022-07-14.json")
```

```python
with open(JSON_FILES[0], "r") as f:
    line = f.readline()
    data = json.loads(line)
    pprint(data)
```

```python
rows=[]
CountDict=[]
for JSON_FILE in JSON_FILES:
    with open(JSON_FILE, "r") as f:
        for line in f:
            data = json.loads(line)
            reply_count = data["reply_count"]
            quote_count = data["quote_count"]
            retweet_count = data["retweet_count"]
            like_count = data["favorite_count"]
            screen_name = data["user"]["screen_name"]
            tweet_id = data["id_str"]
            if "sentiment" in data["entities"]:
                Negative_tweet_sentiment=data["entities"]["sentiment"]["probabilities"]["Very negative"]+data["entities"]["sentiment"]["probabilities"]["Negative"]
                Positive_tweet_sentiment=data["entities"]["sentiment"]["probabilities"]["Very positive"]+data["entities"]["sentiment"]["probabilities"]["Positive"]
            else:
                Negative_tweet_sentiment = None;
                Positive_tweet_sentiment = None;

            if "extended_tweet" in data:
                text = data["extended_tweet"]["full_text"]
            else:
                text = data["text"]
            interact_count = like_count+retweet_count+quote_count+reply_count  
            rows.append((tweet_id, screen_name, text, Negative_tweet_sentiment, Positive_tweet_sentiment, reply_count, quote_count, retweet_count, like_count, interact_count))
```

```python
df = pd.DataFrame(data=rows, columns=["tweet_id", "screen_name", "tweet_text","negative_tweet_sentiment","positive_tweet_sentiment","reply_count","quote_count","retweet_count","like_count","interact_count"])
```
